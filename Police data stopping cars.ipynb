```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline 
import seaborn as sns
```

```python
df = pd.read_csv(r"C:\Users\Kim Yuri\Desktop\GitHub projects\police.csv")
df.head(10)
```

# Race

```python
df['driver_race'].value_counts().plot(kind='pie',autopct='%1.1f%%', figsize=(7, 7))
```

```python
df.info()
```

```python
# Convert the 'stop_time' column to a datetime object
df['stop_time'] = pd.to_datetime(df['stop_time'], format='%M:%S')

# Extract minutes and seconds and convert them to seconds
df['stop_time_seconds'] = df['stop_time'].dt.minute * 60 + df['stop_time'].dt.second
```

```python
df.head(10)
```

```python
df = df.drop('stop_time', axis=1)
```

```python
df.head()
```

```python
test = df.groupby(['driver_race','driver_gender'])['stop_time_seconds'].mean().sort_values(ascending=False)
```

```python
test
```

```python
plot_data = test.unstack()
plot_data
```

```python
plot_data.plot(kind='bar', stacked=True, color=['pink', 'blue'], figsize=(10, 6))
```

# Stop date

```python
df['stop_date'] = pd.to_datetime(df['stop_date'])
df
```

```python
df['year'] = df['stop_date'].dt.year
df['month'] = df['stop_date'].dt.month
df.head()
```

```python
df['year'].value_counts().sort_index().plot(kind='bar')
```

```python
test1 = df.groupby(['year','month']).size()
test2 = test1.unstack()
test2
```

```python
plt.figure(figsize=(20, 14))
```

```python
sns.heatmap(test2, annot=True, cmap='viridis');
```
