# Race

```python
df2 = df1
df2.head()
# make a copy of main df
```

```python
df2['search_conducted'] = df2['search_conducted'].astype(int) #convert search_conducted into int
df2
```

```python
df2['driver_gender'] = df2['driver_gender'].map({'M':1, 'F':0}) #convert driver_gender into int
df2
```

```python
df2['driver_gender'].fillna(1, inplace=True) #I assumed 1 as Male since majority of drivers are male
df2
```

```python
df2['driver_gender'] = df2['driver_gender'].astype(int) 
df2
```

```python
df2['drugs_related_stop'] = df2['drugs_related_stop'].astype(int)  #convert drugs_related_stop into int
df2
```

```python
df2['is_arrested'] = df2['is_arrested'].astype(int)  #convert is_arrested into int
df2
```

### White drivers correlation

```python
white_driver = df2[df2['driver_race'] == 'White']
white_driver
```

```python
corr_cols = ['search_conducted','driver_gender','drugs_related_stop','is_arrested']
```

```python
white_driver = white_driver[corr_cols]
white_driver
```

```python
white_corr = white_driver.corr()
white_corr
```

```python
sns.heatmap(white_corr, cmap="Blues", annot=True);

# low positive correlation between is_arrested and search_conducted (0.35)
# low positive correlation between drug related stop and driver_gender (0.033)
# medium positive strength of correlation between drug_related_stop and if a search is conducted (0.51)
```

```python
white_driver.groupby(['search_conducted','is_arrested']).size().unstack()
# There are 985 White drivers arrested without being searched
```

# Black drivers correlation

```python
black_driver = df2[df2['driver_race'] == 'Black']
black_driver
```

```python
black_driver = black_driver[corr_cols]
black_driver
```

```python
black_corr = black_driver.corr()
black_corr
```

```python
sns.heatmap(black_corr, cmap='Blues', annot=True);

# also medium positive correlation between search_conducted and drug_related_stop (0.49)
# low positive correlation between gender and search_conducted (0.049)
# low positive correlation between drugs_related_stop and is_arrested (0.11)
```

```python
black_driver.groupby(['drugs_related_stop','driver_gender']).size().unstack()

# There are 183 Male black drivers stopped due to drugs
```
